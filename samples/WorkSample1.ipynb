```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_HS = pd.read_csv('homeSales.csv')

df_HS['Date'] = pd.date_range(freq='m', start=pd.Timestamp(year=2007, month=8, day=31), periods = len(df_HS['homeSales']))
df_HS['Year'] = pd.DatetimeIndex(df_HS['Date']).year

fig,ax = plt.subplots()

ax.bar(df_HS['Year'], df_HS['homeSales'])
ax.set_xlabel('Year')
ax.set_ylabel('Units of Demand')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axhline(df_HS['homeSales'].mean(), color='tab:red', linewidth=2, label='Mean Home Sales')
plt.legend()
```
